# For loop to append other prior years

```python
for year in current_yr_lst:
    years_lst = [year - 3, year - 2, year - 1]
        
    print("For years: ", years_lst)
        # temp dataframe with only the years we want
    df_temp = df[df['Year'].isin(years_lst)]
    
    pivot_df_temp = df_temp.pivot_table(values = oput_Cols
                                  ,index = ['Player', 'Max_Team_Name', 'Pos']
                                  ,columns = 'Year'
                                 )
    
    pivot_df_temp['Current_Year'] = year
    
    pivot_df_temp.columns.set_levels(['year - 3', 'year - 2', 'year - 1', ''], level = 1, inplace = True)
    
    pivot_df_full = pd.concat([pivot_df_full, pivot_df_temp])
    
    print("--- Completed ---")
    print("\n")
```

```python
# Create Mapping
qb_mapping = df.query("Pos == 'QB'")
qb_mapping['YdsPerGame'] = qb_mapping['Avg_PassingYds']
qb_mapping['TDPerGame'] = qb_mapping['Avg_PassingTD']
qb_mapping['YdsPerAttempt'] = qb_mapping['YardsPerAttempt']
qb_mapping['Run_Percentage'] = qb_mapping['Team_RushingAtt'] / (qb_mapping['Team_RushingAtt'] + qb_mapping['Team_Tgt'])
qb_mapping = qb_mapping[['Player','Team_Name','Pos','Year','Weeks_Played','YdsPerGame','TDPerGame','YdsPerAttempt','Run_Percentage']]


# Team Mapping for RB 
team_mapping = df.query("Pos == 'RB'")
team_mapping['Rushing_YdsPerAttempt'] = team_mapping['Team_RushingYds'] / team_mapping['Team_RushingAtt']
team_mapping['YdsPerGame'] = team_mapping['Team_RushingYds']
team_mapping['TDPerGame'] = team_mapping['Team_RushingTD']
team_mapping['Rec_YdsPerAttempt'] = team_mapping['Team_ReceivingYds'] / team_mapping['Team_Tgt']
team_mapping['Run_Percentage'] = team_mapping['Team_RushingAtt'] / (team_mapping['Team_RushingAtt'] + team_mapping['Team_Rec'])
team_mapping = team_mapping[['Team_Name','Year','Team_RushingYds','Team_RushingTD','Rushing_YdsPerAttempt','Team_ReceivingYds','Team_ReceivingTD','Run_Percentage']]
team_mapping = team_mapping.drop_duplicates()


qb_mapping.head()

# # Team Mapping for Receivers
# rec_mapping = df.query("Pos == 'WR' or Pos == 'TE'")
# rec_mapping['YdsPerGame'] = rec_mapping['Avg_ReceivingYds']
# rec_mapping['TDPerGame'] = rec_mapping['Avg_ReceivingTD']
# rec_mapping['YdsPerAttempt'] = rec_mapping['YardsPerAttempt']
# rec_mapping['Run_Percentage'] = rec_mapping['Sum_RushingAtt'] / (rec_mapping['Sum_RushingAtt'] + rec_mapping['Sum_Rec'])
# rec_mapping = rec_mapping[['Player','Max_Team_Name','Pos','Year','YdsPerGame','TDPerGame','YdsPerAttempt','Run_Percentage']]

# Union all dataframes
#countepart_mapping = pd.concat([qb_mapping, rb_mapping, rec_mapping])
#countepart_mapping.head()
```

```python
# Update Max Team Name for 2022 using trades file
trades_df = pd.read_excel('/content/gdrive/MyDrive/DVA Project/Data/trade_data.xlsx')
```

```python
# Update Mapping for Max Team
pivot_df_full = pd.merge(pivot_df_full, trades_df, right_on=['Trade_Year','Player','Pos'], left_on=['Current_Year','Player','Pos'], how='left',  suffixes=('', '_DROP')).filter(regex='^(?!.*_DROP)')
pivot_df_full['New_Team'] = pivot_df_full['New_Team'].fillna(pivot_df_full['Max_Team_Name'])
pivot_df_full['Max_Team_Name'] = pivot_df_full['New_Team']
pivot_df_full.drop(['Trade_Year','New_Team'], axis=1, inplace=True)
```

```python
# Read in Data
qb_stats = pd.read_excel('/content/gdrive/MyDrive/DVA Project/Data/qb_stats.xlsx', index_col=[0])

# Make copy of data
rb_df = pivot_df_full[pivot_df_full['Pos'] != 'QB']
qb_df = pivot_df_full[pivot_df_full['Pos'] == 'QB']

# Join qb_stats with rb_df
rb_df = pd.merge(rb_df, qb_stats, left_on=['Current_Year','Max_Team_Name'], right_on=['Join_Year','Team_Name'], suffixes=('','_QB'), how='left')
rb_df = rb_df.drop(columns=['Player_QB','Join_Year', 'Year', 'Team_Name'])
player_plus_team_v2 = pd.concat([rb_df, qb_df])
player_plus_team_v2.fillna(value=0, inplace=True)

player_plus_team_v2
```
