```python
print(df['Year'].min())
print(df['Year'].max())
```

```python
col_lst = df.columns.tolist()
print(len(col_lst))


remove_Cols = ['Unnamed: 0', 'Player', 'Pos', 'Team_Name', 'Year']
oput_Cols = [item for item in col_lst if item not in remove_Cols]
print(len(oput_Cols))
```

# Creating initial dataframe for 2022 as current year

```python
current_year = 2022
years_lst = [2019, 2020, 2021]
df_temp = df[df['Year'].isin(years_lst)]

print(df_temp['Year'].unique())

pivot_df_full = df_temp.pivot_table(values = oput_Cols
                              ,index = ['Player', 'Team_Name', 'Pos']
                              ,columns = 'Year'
                             )

pivot_df_full['Current_Year'] = current_year

    # renaming the columns to be able to concatenate them later!
pivot_df_full.columns.set_levels(['year - 1', 'year - 2', 'year - 3', ''], level = 1, inplace = True)

pivot_df_full.head()
```

```python
pivot_df_full['Current_Year'].unique()
```

```python
pivot_df_full.columns.levels[1]
```

# For loop to append other prior years

```python
for year in current_yr_lst:
    years_lst = [year - 3, year - 2, year - 1]
        
    print("For years: ", years_lst)
        # temp dataframe with only the years we want
    df_temp = df[df['Year'].isin(years_lst)]
    
    pivot_df_temp = df_temp.pivot_table(values = oput_Cols
                                  ,index = ['Player', 'Team_Name', 'Pos']
                                  ,columns = 'Year'
                                 )
    
    pivot_df_temp['Current_Year'] = year
    
    pivot_df_temp.columns.set_levels(['year - 1', 'year - 2', 'year - 3', ''], level = 1, inplace = True)
    
    pivot_df_full = pd.concat([pivot_df_full, pivot_df_temp])
    
    print("--- Completed ---")
    print("\n")
```
